```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# Load the dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#cleaning the data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US= df[df['job_country']== 'United States'].copy()
df_US
```

```python
df_US['job_posted_month']= df_US['job_posted_date'].dt.strftime('%B')
df_US
```

```python
df_US_pivot =df_US.pivot_table(
    index='job_posted_month',
    columns='job_title_short',
    aggfunc= 'size')
df_US_pivot
```

```python
df_US_pivot.drop(columns='index', inplace= True)
```

```python
df_US_pivot.sort_values('month no', inplace=True)
df_US_pivot
```

```python
df_US_pivot.set_index('job_posted_month', inplace=True)
df_US_pivot
```

```python
df_US_pivot.drop(columns='month no', inplace=True)
df_US_pivot
```

```python
top_3 = df_US['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()
top_3
```

```python
df_US_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Demand of Top Data jobs in the US')
plt.xlabel('')
plt.ylabel('Number of Job Postings')
plt.legend(title='')
plt.show()
```
