```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# Load the dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#cleaning the data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df
```

```python
df_india = df[df['job_country'] == 'India']
df_india
```

```python
df_india.reset_index(inplace=True)
df_india
```

```python
df_usa = df[df['job_country'] == 'United States']
df_usa.reset_index(inplace=True)
df_usa.set_index('job_index', inplace=True)
df_usa
```

```python
median_pivot_usa = df_usa.pivot_table(
    values='salary_year_avg',
    index='job_title_short',
    aggfunc=['median', 'min', 'max'])
median_pivot_usa
```

```python
median_pivot_usa.sort_values(by=('median', 'salary_year_avg'), ascending=False)
```
