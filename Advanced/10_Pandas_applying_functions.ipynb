```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# Load the dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#cleaning the data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_salary = df[pd.notna(df['salary_year_avg'])].copy()
```

```python
def projected_salary(salary):
    return salary * 1.04

df_salary['salary_next_year']= df_salary['salary_year_avg'].apply(projected_salary)
df_salary[['salary_year_avg', 'salary_next_year']].head(10)
```

If senior then salary increased by 5%  else increased by 3%

```python
def projected_salary(row):
    if "Senior" in row['job_title_short']:
        return 1.05 * row['salary_year_avg']
    else:
        return 1.03 * row['salary_year_avg']
```

```python
df_salary['salary_next_year']= df_salary.apply(projected_salary, axis=1)
df_salary
```
