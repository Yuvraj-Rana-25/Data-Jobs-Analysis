```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# Load the dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#cleaning the data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_extra = df.copy()
df_extra
```

```python
df_extra['job_posted_month']= df_extra['job_posted_date'].dt.strftime('%b')
df_extra
```

```python
months = df_extra['job_posted_month'].unique().tolist()
months 
```

```python
dict_months = {month : df_extra[df_extra['job_posted_month'] == month] for month in months}
dict_months['Jan']
```

```python
df_q1 = pd.concat([dict_months['Jan'], dict_months['Feb'], dict_months['Mar']], ignore_index=True)

df_q1['job_posted_month'].value_counts().plot(kind='bar')
plt.xticks(rotation=0)
plt.title('Job Postings in Q1')
plt.xlabel('')
plt.ylabel('Number of job postings')
plt.show()
```

# Pandas Dataframe Exporting

```python
df_q1.to_csv('job_postings_q1.csv')
```
